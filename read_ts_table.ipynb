```python
import tstables as ts
import tables
import pandas as pd
from datetime import datetime,timedelta
import h5py
```

# 读取交易员交易记录

```python
path = 'd:/datas/cta_tstable.h5'

def get_tstable(ins, start_dt, end_dt):
    with tables.open_file(path, 'r') as f:
        instrument = '/'+ins
        if f.__contains__(instrument):
            ts = f.get_node(instrument)._f_get_timeseries()
            return ts.read_range(start_dt, end_dt)
    return pd.DataFrame()

start_dt = datetime(2016, 12, 5)
end_dt = datetime(2017, 7, 31, 15, 0, 0)
df = get_tstable('a1709', start_dt, end_dt)
df
```

# 读取Tick数据

```python
def get_tstable(ins, start_dt, end_dt):
    with tables.open_file('d:/datas/ts_futures.h5', 'r') as f:
        for market in ['dc', 'sc', 'zc']:
            instrument = '/{}/{}'.format(market, ins)
            print(instrument)
            if f.__contains__(instrument):
                ts = f.get_node(instrument)._f_get_timeseries()
                return ts.read_range(start_dt, end_dt)
                    
    return pd.DataFrame()

start_dt = datetime(2016, 12, 5)
end_dt = datetime(2017, 7, 31, 23, 59, 59)
df = get_tstable('cu1705', start_dt, end_dt)
df
```
