```python
def get_next_month(dt):
    dt = pd.to_datetime(dt)
    a = dt.year
    b = dt.month + 1 
    c = dt.day
    new_dt = '%d-%d-%d'%(a,b,c)
    new_dt = pd.to_datetime(new_dt)
    return new_dt
    
```

```python
x = '2016-01-30'
get_next_month(x)
```
