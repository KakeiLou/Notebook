```python
import pandas as pd
from datetime import datetime
df = pd.read_hdf('C:/Users/Administrator/Desktop/Work/data/CTA_trade_detail_all/trade_detail_007738030022.h5')
```

```python
df_expands = []
def expand_by_lots(row):
    lots = row.lots
    row.lots = 1
    df_expands.append(pd.DataFrame([row for x in range(lots)]))
    return row

df.apply(expand_by_lots, axis=1)
df_expand = pd.concat(df_expands[1:])
```
