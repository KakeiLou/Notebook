```python
import pandas as pd
import numpy as np
import phoenix_rtn3 as pho
import os
from datetime import datetime
from dateutil.relativedelta import relativedelta
%matplotlib inline
```

```python
path = 'c:/Users/Administrator/Desktop/Work/Projects/Autocall/price_sz'
dfs = []
stock = []
for f in os.listdir(path):
    if f.find('.csv') == -1:
        continue
    df = pd.read_csv(os.path.join(path, f), index_col=0, parse_dates=True,
                    header= 0, names =['date','close'])
    if len(df) == 449:
        stock.append(f[:f.find('.')])
        dfs.append(df)
```

```python
def find_result(df_1,df_2, rtn_rate, lowerbond):
    rtn_list = []
    for i in df_1.index:
        r = pho.phoenix_rtn (df_2, i, 6, rtn_rate, lowerbond)
        rtn_list.append(r)
    sr1 = pd.Series(rtn_list)
    m = sr1.mean()
    #print ('Avg profit of this product is %f'%m)
    p = sr1[sr1>0].count()/sr1.count()
    #print ('Probability of winning is  %f'%p)
    return [m,p]
```

```python
dfss = []
for i in dfs:
    df1 = i[0:325]
    dfss.append(df1)

rtn=[]
winrate=[]
for i in range(len(dfss)):
    m = find_result(dfss[i], dfs[i],0.025, 0.9)
    rtn.append(m[0])
    winrate.append(m[1])
df_rtn = pd.DataFrame(rtn)
df_win = pd.DataFrame(winrate)
df_stock = pd.DataFrame(stock)
df_total = pd.concat([df_stock,df_rtn, df_win], axis=1)
df_total.columns = ['stock','rtn','winrate']
#df_total.to_excel('sz50_stocks.xlsx')
df_total
```
