```python
def find_result(df_1,df_2, rtn_rate, lowerbond):
    rtn_list = []
    for i in df_1.index:
        r = pho.phoenix_rtn (df_2, i, 6, rtn_rate, lowerbond)
        rtn_list.append(r)
    sr1 = pd.Series(rtn_list)
    m = sr1.mean()
    #print ('Avg profit of this product is %f'%m)
    p = sr1[sr1>0].count()/sr1.count()
    #print ('Probability of winning is  %f'%p)
    return [m,p]
```

```python
find_result(df2,df1, 0.025, 0.9)
```
