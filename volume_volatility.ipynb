```python
df2 = pd.read_csv('C:/Users/Administrator/Desktop/Work/data/CTA_algo/result/m1709_mins.csv')
df2
```

```python
df2.columns = ['datetime','close','day']
```

```python
df2.index = df2.day
df2
```

```python
df2['ln_rtn'] = (df2.close / df2.close.shift(1)).apply(math.log)
df2.index = pd.to_datetime(df2.index)
```

```python
df2.groupby(df2.index).apply(lambda x :x[x.close!=0])
```

```python
sr1 = df2.groupby(df2.index).ln_rtn.std()
df3  = pd.DataFrame(sr1)
df3['annual_vol'] = df3.ln_rtn * math.sqrt(244)
df3
```

```python
df3.to_excel('m1709_dayli_vol.xlsx')
```
