```python
import pandas as pd
```

```python
df = pd.read_hdf('C:\\Users\\Administrator\\Desktop\\mergehsi\\merge_HSI.h5')
df
```

```python
sr1 = df.resample('5min').vol.sum()
sr1
```

```python
sr2 = df.resample('5min').price.first()
sr2
```

```python
sr3 = df.resample('5min').price.last()
sr3
```

```python
sr4 = df.resample('5min').price.max()
sr4
```

```python
sr5 = df.resample('5min').price.min()
sr5
```

```python
df1 = pd.DataFrame([sr1,sr2,sr3,sr4,sr5])
df1 = df1.T
df1 
```

```python
df1.columns = (['volume','first_price','last_price','max_price', 'min_price'])
```

```python
df1
```

```python
df1.to_hdf('5min_hsidata.h5','HSI')
```

```python
df1['rtn'] = (df1.last_price - df1.last_price.shift(1))/df1.last_price.shift(1)
```

```python
from math import log
```

```python
df1['log'] = df1.last_price.apply(lambda x: log(x))
```

```python
df1['logRtn'] = df1.log - df1.log.shift(1)
```

```python
df1
```

```python
df1.drop('log',axis=1,inplace=True)
df1
```

```python
df1.to_hdf('5min_hsidata.h5','hsi')
```
